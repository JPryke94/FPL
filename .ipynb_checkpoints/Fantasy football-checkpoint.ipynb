```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns

```

```python
def get_data(x,y,z):
    
    # Get data for gameweeks and teams
    gw1 = pd.read_csv(f"/Users/pryke/Documents/Projects/Fantasy football/Data/{x}.csv")
    gw2 = pd.read_csv(f"/Users/pryke/Documents/Projects/Fantasy football/Data/{y}.csv")
    gw3 = pd.read_csv(f"/Users/pryke/Documents/Projects/Fantasy football/Data/{z}.csv")
    teams = pd.read_csv(f"/Users/pryke/Documents/Projects/Fantasy football/Data/teams.csv")
    
    # rename columns in teams to align with gameweeks
    teams.rename(columns= {'id':'opponent_team','name':'opponent_team_name'}, inplace= True)
    
    return gw1,gw2,gw3,teams
```

```python
gw1,gw2,gw3,teams = get_data('gw1','gw2','gw3')
```

```python
def clean_data(week1=gw1,week2=gw2,week3=gw3,teams=teams):
    
    week = 1

    for x in [week1,week2,week3]:
        x['round'] = week
        x.drop(labels = ['xP','fixture','selected','team_a_score','team_h_score','transfers_balance','transfers_in','transfers_out'],axis=1,inplace=True)
        week+=1
    
    gw1 = week1.merge(teams,on='opponent_team')
    gw2 = week2.merge(teams,on='opponent_team')
    gw3 = week3.merge(teams,on='opponent_team')
    
    return gw1,gw2,gw3
```

```python
gw1,gw2,gw3 = clean_data()
```

```python
def create_database(gw1=gw1,gw2=gw2,gw3=gw3):
    
    data = gw1.append(other = gw2, ignore_index=True)
    data = data.append(other = gw3, ignore_index=True)
    
    return data
```

```python
data = create_database()
```

```python
data.shape
```

```python
def calculate_averages(data=data):
    three_week_average = round(data.groupby('name').mean(),3)
    three_week_average = three_week_average.rename(columns=lambda x: 'avg_'+x)
    return three_week_average
```

```python
three_week_average = calculate_averages()
```
